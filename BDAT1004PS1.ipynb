```python
##so I used the following site to figure out what are the vowles and consonant letters. Hope that is Okay!
##https://teachphonics.blog/2019/10/10/what-are-vowels-and-consonants/


cons = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k','l','m','n','p','q','r','s','t','v','w','x','y','z']
vowels = ['a', 'e', 'i', 'o', 'u']


def pig(wrd):
    wrd = wrd.lower()
    if(wrd[0] in cons):
        wrd = wrd[1:] + wrd[0] + "ay"
        return wrd
    
    
    if(wrd[0] in vowels):
        wrd = wrd + "way"
        return wrd

print(pig("happy"))
print(pig("Enter"))
```

```python
def bldcount(fileName):
    
    infile = open(fileName, 'r+')
    lines = infile.readline().upper().strip().split(" ")
    bldTypes = ['A','B','AB','O','OO']
    for bld in bldTypes:
        print("There are {0} patients of blood type {1}.".format(lines.count(bld), bld))
    infile.close()
    
bldcount('bloodtype1.txt')
```

## Q8

```python
def curconv(currency, amount):
    infile = open('currencies.txt', 'r+')
    lines = infile.readlines()
    for i in range(len(lines)):
        if currency  in lines[i].split():
            return amount * float(lines[i].split()[1])
            
        
    
curconv('EUR', 10)
```

```python
def  frequencies(text):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    lst =[]
    for i in letters:
        lst.append(text.count(i))
    return lst
print(frequencies("The quick red fox got bored and went home."))

print(frequencies("apple"))
```
